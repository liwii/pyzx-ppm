```python
from ppm import *
from util import *
```

```python
lanes = 6
pattern = block_pattern(6, 10)
```

```python
pattern
```

```python
ppm = PPM(lanes)
for (p, (g1, g2, g3)) in pattern:
    ppm.add_gate(g1, p)
    ppm.add_gate('HSH', p)
    ppm.add_gate('HSH', p + 1)
    ppm.add_gate(g2, p + 1)
    ppm.add_gate(g3, (p, p + 1))
    ppm.add_gate('HSH', p)
    ppm.add_gate('HSH', p + 1)
```
