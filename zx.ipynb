```python
import pyzx as zx
```

```python
qubit_amount = 5
gate_count = 80
#Generate random circuit of Clifford gates
circuit = zx.generate.cliffordT(qubit_amount, gate_count)
```

```python
circuit
```

```python
zx.draw(circuit)
```

```python
c_original = zx.extract_circuit(circuit.copy())
```

```python
print(c_original.stats())
```

```python
zx.simplify.full_reduce(circuit)
```

```python
zx.draw(circuit)
```

```python
c_opt = zx.extract_circuit(circuit.copy())
```

```python
c_opt
```

```python
print(c_opt.stats())
```

```python
g = c_opt.to_graph()
```

```python
zx.draw(g)
```

```python
c_opt.gates
```

```python
c_basic = c_opt.to_basic_gates()
```

```python
print(c_basic.stats())
```

```python
c_basic.gates
```

```python
zx.draw(c_opt)
```
